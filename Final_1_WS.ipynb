# WebScraping

```python
link = []
titulo = []
autor = []
ws_paginas = []

ws_generos = []

ws_nota = []
ws_leram = []
ws_favoritos = []

ws_nota_renata = []
ws_ano_leitura = []

for livro in renta_lidos:
    url_livro = livro
    response_livro = requests.get(url_livro)
    html_livro = response_livro.content
    soup_livro = BeautifulSoup(html_livro)
    
    link += [livro]
    titulo += [soup_livro.find_all(['strong'])[1].string]
    try:
        autor += [soup_livro.find_all(['h3'])[-1].string]
    except:
        autor += ["-"]
    try:
        ws_paginas += [soup_livro.find_all(['div'], attrs = {"class":"sidebar-desc"})[0].contents[6][-3:]]
    except:
        ws_paginas += ["0"]
    
    try:
        ws_generos += [soup_livro.find_all(['span'], attrs = {"class" : "pg-livro-generos"})[0].text]
    except:
        ws_generos += ["-"]
    
    ws_nota += [soup_livro.find_all(['star-rating'], attrs = {"size":"medium"})[0].get("rate")]
    ws_leram += [soup_livro.find_all(['b'])[6].text]
    ws_favoritos += [soup_livro.find_all(['div','a'], attrs = {"class":"link_11_off l_icones sk-cor-branco"})[0].text.split(" ")[-1]]
    
    #ws_nota_renata += [soup_livro.find_all(['star-rating'])[0].get("rate")]
    
    #try:
     #   ws_nota_renata += [soup_livro.find_all(['star-rating'])[0].get("rate")]
    #except:
     #   ws_nota_renata += [soup_livro.find_all(['star-rating'])[0].get("rate")]
    #ws_ano_leitura += []
```

```python
paginas = [int(i.replace(".","")) for i in ws_paginas]

nota = [float(n) for n in ws_nota]
leram = [int(i.replace(".","")) for i in ws_leram]
favoritos = [int(i.replace("(","").replace(")","").replace(".","")) for i in ws_favoritos]

nota_renata = [float(n) for n in ws_nota_renata]
ano_leitura = [float(n) for n in ws_ano_leitura]

df_renta_lidos = pd.DataFrame({"titulo" : titulo, 
                               "link" : link, 
                               "autor" : autor, 
                               "paginas" : paginas, 
                               "genero" : ws_generos, 
                               "nota" : nota,
                               "leram" : leram,
                               "favoritos" : favoritos,
                               #"nota_renata" : nota_renata,
                               #"ano_leitura" : ano_leitura
                              })
df_renta_lidos
```
